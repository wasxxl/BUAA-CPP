```python
import pandas as pd

df = pd.read_csv('dataset/C2022/total.csv')
```

```python
print(len(df['language']))
print((df['language']=='c').sum())
print((df['language']!='c').sum())
```

```python
question_info = pd.read_excel('dataset/C2022/question_info_meta.xlsx')

question_info.head(10)
```

```python
question_info.columns
```

```python
def get_xth_alphabet_upper(x):
    if x < 1 or x > 26:
        return "Invalid input. Please provide a number between 1 and 26."
    return chr(x + 96).upper()

question_info.columns = ['qid', 'qid_text', 'pass/user_num', 'pass/submission_num', 'problem_set', 'problem_name', 'tag1',
       'tag2', 'tag3']

question_info.head(10)
question_info['problem_name'] = question_info['problem_set'] + '_' + question_info['problem_name'].apply(get_xth_alphabet_upper)
```

```python
df = df.merge(question_info, on=['problem_set', 'problem_name'])
```

```python
df.head(10)
```

```python

tag_set = set()
for col in ['tag1','tag2','tag3']:
       tag_set.update(df[col].unique())

tag_set
```

```python
import tokenize
from io import BytesIO
def calculate_line(code_text):
    code_lines = code_text.split('\n')
    line_count = len(code_lines)

    token_count = 0
    for line in code_lines:
        try:
            # Tokenize the line and count tokens
            token_generator = tokenize.tokenize(BytesIO(line.encode('utf-8')).readline)
            for _ in token_generator:
                token_count += 1
        except tokenize.TokenError:
            pass

    return line_count
def calculate_token(code_text):
    code_lines = code_text.split('\n')
    line_count = len(code_lines)

    token_count = 0
    for line in code_lines:
        try:
            # Tokenize the line and count tokens
            token_generator = tokenize.tokenize(BytesIO(line.encode('utf-8')).readline)
            for _ in token_generator:
                token_count += 1
        except tokenize.TokenError:
            pass

    return token_count
```

```python

df['code_lines'] = df['code'].apply(calculate_line)
df['token_nums'] = df['code'].apply(calculate_token)
```

```python
print(df['code_lines'].describe())
import matplotlib.pyplot as plt
# Plotting the score distribution
plt.figure(figsize=(8, 6))
plt.hist(df['code_lines'],bins=100,density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('code lines distribution')
plt.xlabel('line number')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

```python
print(df['token_nums'].describe())
import matplotlib.pyplot as plt
# Plotting the score distribution
plt.figure(figsize=(8, 6))
plt.hist(df['token_nums'],bins=100,density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('token nums distribution')
plt.xlabel('line number')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```
