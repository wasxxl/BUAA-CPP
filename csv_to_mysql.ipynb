```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

# Function to create a connection to the MySQL database
def create_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as e:
        print(f"The error '{e}' occurred")
    return connection

# Function to execute a query
def execute_query(connection, query, data):
    cursor = connection.cursor()
    try:
        cursor.executemany(query, data)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")

# Database connection parameters
host_name = "localhost"
user_name = "root"
user_password = "17377756"
db_name = "buaa_online_judge"

# Create a connection to the database
connection = create_connection(host_name, user_name, user_password, db_name)
```

```python
def select_max_primary_key(table_name):
    connection = create_connection(host_name, user_name, user_password, db_name)
    cursor = connection.cursor()

    query = f"SHOW COLUMNS FROM {table_name}"
    cursor.execute(query)

    # 获取表结构
    columns = cursor.fetchall()

    # 查找主键列
    pri_key_column = None
    print(columns)
    for column in columns:
        print(column[3])
        if column[3] == 'PRI':
            pri_key_column = column[0]
            break

    # 如果找到了主键列，查询最大值
    if pri_key_column:
        max_id_query = f"SELECT MAX({pri_key_column}) FROM {table_name}"
        cursor.execute(max_id_query)
        result = cursor.fetchone()
        max_id = result[0] if result else None
        print(f"The maximum {pri_key_column} value is:", max_id)
        return max_id
    else:
        print("No primary key found in the table.")


max_relation_id = select_max_primary_key('course_student_relationships')
```

```python
from datetime import datetime, timedelta
def get_next_time(time_str, seconds):
    # Parse the input time string to a datetime object
    dt = datetime.strptime(time_str, '%Y-%m-%d %H:%M:%S')

    # Add the specified number of seconds
    new_dt = dt + timedelta(seconds=seconds)

    # Format the new datetime object back to a string
    return new_dt.strftime('%Y-%m-%d %H:%M:%S')
```
